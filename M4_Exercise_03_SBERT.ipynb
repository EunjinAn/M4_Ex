```python
from sentence_transformers import SentenceTransformer, util
import torch
import numpy as np
from sentence_transformers.util import cos_sim
import gradio as gr
```

```python
# sentences = ["This is an example sentence", "Each sentence is converted"]

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
model

# embeddings = model.encode(sentences)
# print(embeddings)
```

```python
sentences = ['I do not know, this is sucess or not',
            'I tried SBERT, but I am not sure what I am doing',
            'this country is the happist county.',
            'Denmark is the happist country.',
            'I am happy']
```

```python
embeddings = model.encode(sentences)
embeddings.shape
```

```python
similarity = np.zeros((len(sentences),len(sentences)))

for i in range(len(sentences)):
  similarity[i:, i] = cos_sim(embeddings[i], embeddings[i:])
similarity
```

```python
den_sentences = ['Denmark is a Nordic country located in Northern Europe.',
'The official language of Denmark is Danish.',
'The capital city of Denmark is Copenhagen.',
'Denmark has a population of approximately 5.8 million people.',
'The Danish flag is known as the Dannebrog.',
'Denmark is a constitutional monarchy, with Queen Margrethe II as the current monarch.',
'Denmark is known for its high quality of life and is consistently ranked as one of the happiest countries in the world.',
'Danish cuisine includes dishes like smørrebrød (open-faced sandwiches) and frikadeller (meatballs).',
'Danish design is renowned for its clean, simple lines and functionalism.',
'Hans Christian Andersen, the author of classic fairy tales such as "The Little Mermaid" and "The Ugly Duckling," was Danish.',
'The Danish Royal Family is popular both domestically and abroad.',
'Denmark is the southernmost of the Nordic countries.',
'Danish culture is known for its emphasis on hygge, or coziness and contentment.',
'The city of Aarhus in Denmark was named the European Capital of Culture in 2017.',
"Denmark is a leader in renewable energy, with wind power providing a significant portion of the country's energy needs.",
'Denmark has a strong tradition of bicycling and is considered one of the most bike-friendly countries in the world.',
'Legoland, a popular amusement park dedicated to the iconic toy bricks, is located in Denmark.',
'The Danish National Museum in Copenhagen is home to a vast collection of artifacts and exhibits on Danish history and culture.',
'The Danish concept of "janteloven" refers to a cultural norm that discourages individualism and emphasizes modesty and equality.',
'Denmark has a high standard of living and a robust social welfare system.',
'The Danish language is closely related to Norwegian and Swedish.',
'Denmark is home to many picturesque villages and charming seaside towns.',
'The Viking Age began in Denmark, and many important Viking artifacts have been discovered in the country.',
'The Danish fashion brand, Ganni, has gained international acclaim in recent years.',
'The Øresund Bridge connects Denmark to Sweden, spanning 8 kilometers across the Øresund Strait.',
'Denmark is known for its high-quality design and architecture, with buildings like the Sydney Opera House and the Shard in London designed by Danish architects.']
```

```python
corpus_embeddings = model.encode(den_sentences, convert_to_tensor=True)


# Find the closest 5 sentences of the corpus for each query sentence based on cosine similarity
top_k = min(5, len(den_sentences))

for query in sentences:
    query_embedding = model.encode(query, convert_to_tensor=True)

    # We use cosine-similarity and torch.topk to find the highest 5 scores
    cos_scores = util.cos_sim(query_embedding, corpus_embeddings)[0]
    top_results = torch.topk(cos_scores, k=top_k)

    print("\n\n======================\n\n")
    print("Query:", query)
    print("\nTop 5 most similar sentences in corpus:")

    for score, idx in zip(top_results[0], top_results[1]):
        print(den_sentences[idx], "(Score: {:.4f})".format(score))
```

```python
def query_corpus(query):
  results = []
  query_embedding = model.encode(query, convert_to_tensor=True)
  cos_scores = util.cos_sim(query_embedding, corpus_embeddings)[0]
  top_results = torch.topk(cos_scores, k=top_k)
  for score, idx in zip(top_results[0], top_results[1]):
    results.append((den_sentences[idx], score))
    output = results[0][0]
    score = float(results[0][1].numpy())
  return output, score
```

```python
query_corpus('people who live in Aalborg are happy.')
```

```python
import gradio as gr


gr.Interface(fn=query_corpus, inputs=["text"], outputs=["text", "label"]).launch()
```

```python
from datasets import load_dataset

# download dataset from Hugging face
data = load_dataset(
    'jamescalam/youtube-transcriptions',
    split='train'
)

new_data = []  # this will store adjusted data

window = 6  # number of sentences to combine
stride = 3  # number of sentences to 'stride' over, used to create overlap

for i in range(0, len(data), stride):
    i_end = min(len(data)-1, i+window)
    if data[i]['title'] != data[i_end]['title']:
        # in this case we skip this entry as we have start/end of two videos
        continue
    # create larger text chunk
    text = ' '.join(data[i:i_end]['text'])
    # add to adjusted data list
    new_data.append({
        'start': data[i]['start'],
        'end': data[i_end]['end'],
        'title': data[i]['title'],
        'text': text,
        'id': data[i]['id'],
        'url': data[i]['url'],
        'published': data[i]['published']
    })
```

```python
import datasets
import pandas as pd
import numpy as np
```

```python
# we selected train data(100) and test data(10)
train_dataset = datasets.load_dataset('jamescalam/youtube-transcriptions', split="train[:100]")
test_dataset = datasets.load_dataset('jamescalam/youtube-transcriptions', split="train[100:110]")

df_test = pd.DataFrame(test_dataset)
df_train = pd.DataFrame(train_dataset)
```

```python
# model 'AI-Growth-Lab/PatentSBERTa'
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('AI-Growth-Lab/PatentSBERTa')

#Sentences are encoded by calling model.encode()
claim_embeddings = model.encode(df_train.text[:100], convert_to_tensor=True, show_progress_bar=True)
stored_patent_train_embeddings_id = df_train['id']
```

```python
def get_top_n_similar_patents_df(new_claim):
    search_hits_list = []
    new_claim_embedding = model.encode(str(new_claim), convert_to_tensor=True)
    search_hits = util.semantic_search(new_claim_embedding, claim_embeddings, 10000, 5000000, 20)
    top_claim_order = []
    top_claim_ids = []
    top_claim_text = []
    top_similarity_scores = []
    train_text_list = df_train.text.to_list()
    for item in range(len(search_hits[0])):
        top_claim_order = search_hits[0][item].get('corpus_id')
        top_claim_ids.append(stored_patent_train_embeddings_id[top_claim_order])
        top_similarity_scores.append(search_hits[0][item].get('score'))
        top_claim_text.append(train_text_list[top_claim_order][3:])
        
    top_100_similar_patents_df = pd.DataFrame({
        'top_claim_ids': top_claim_ids,
        'claim':top_claim_text,
        'cosine_similarity': top_similarity_scores,
    })

        
    
    return top_100_similar_patents_df
```

```python
title = "Patent Search using PatentSBERTa"

demo = gr.Interface(fn=get_top_n_similar_patents_df, inputs="text", outputs="dataframe")
    
if __name__ == "__main__":
    demo.launch()   
```
